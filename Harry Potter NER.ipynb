```python
import nltk
import time
import nltk.tokenize
```

```python

def findcast(url):
    with open(url,'r') as myplot:
        plot = myplot.read().replace('\n',' ').split('.') # replacing the new line with space and then seperating the sentences
        
    hp_cast = []
    hp_char = []
    hp_char1 = []
    op = []
    output = []
    try:
        for item in plot:
            tokenized = nltk.word_tokenize(item)
            tagged = nltk.pos_tag(tokenized)
            #print tokenized
            
            NamedEnt = nltk.ne_chunk(tagged, binary = False)
            
            
            hp_cast.append([' '.join([y[0] for y in x.leaves()]) for x in NamedEnt.subtrees() if x.label() == "PERSON"])
        
        #hp_char1.append(char for sublist in hp_cast for char in sublist)
        
        for sublist in hp_cast:
            for char in sublist:
                hp_char.append(char)
                
        #op.append(','.join([item for item in hp_char if item not in output]))
        
        for item in hp_char:
            if item not in output:
                output.append(item)
        
    except Exception,e:
        print str(e)
    
    #print output
    #print "The cast of "+url+" is as follows:\n\n",output[5],len(output)
    for character in output:
        print character
    #print hp_cast_set
    #print "\n\nhp char 1:\n\n",hp_char1[0]
    #print hp_char1
    #print op
    
```
